```python
import pandas as pd
import plotly.express as px

# Carregamento com parsing otimizado de datas
df = pd.read_csv('../data/raw/train.csv', parse_dates=['date'])

print("=== INFORMAÇÕES BÁSICAS ===")
print(f"Shape do dataset: {df.shape}")
print(f"Período: {df['date'].min()} a {df['date'].max()}")
print(f"Número de lojas: {df['store'].nunique()}")
print(f"Número de itens: {df['item'].nunique()}")

print("\n=== ESTATÍSTICAS DESCRITIVAS ===")
print(df.describe())
```

```python
# Sazonalidade
fig = px.line(df.groupby('date')['sales'].sum().reset_index(), 
              x='date', y='sales', title='Vendas Totais ao Longo do Tempo')
fig.show()
```

```python
# Distribuição por item/store
fig = px.box(df, x='item', y='sales', title='Distribuição de Vendas por Item')
fig.show()
```

```python
# Análise de sazonalidade
df['day_of_week'] = df['date'].dt.day_name()
df['month'] = df['date'].dt.month

# Padrão semanal
weekly_pattern = df.groupby('day_of_week')['sales'].mean().reindex([
    'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'
])

fig_weekly = px.bar(
    x=weekly_pattern.index, 
    y=weekly_pattern.values,
    title='Padrão Semanal Médio de Vendas'
)
fig_weekly.show()
```
